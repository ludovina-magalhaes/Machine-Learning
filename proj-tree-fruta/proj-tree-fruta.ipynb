```python
import pandas as pd
import numpy as np
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
# Carregar os dados
df = pd.read_excel(r"C:\Users\lagma\Desktop\projetos\machine-learning\data\dados_frutas.xlsx")
print(df)
```

```python
# Aplicar filtros para descobrir a fruta
filtro_redonda = df['Arredondada'] == 1
filtro_suculenta = df['Suculenta'] == 1
filtro_vermelha = df['Vermelha'] == 1
filtro_doce = df['Doce'] == 1

print(df[filtro_redonda & filtro_suculenta & filtro_vermelha & filtro_doce])
```

```python
# Treinar a árvore de decisão
features = ['Arredondada', 'Suculenta', 'Vermelha', 'Doce']
target = 'Fruta'

X = df[features]
y = df[target]

arvore = tree.DecisionTreeClassifier(random_state=42)
arvore.fit(X, y)
```

```python
# Visualizar a árvore de decisão
plt.figure(dpi=600)
tree.plot_tree(arvore,
               class_names=arvore.classes_,
               feature_names=features,
               filled=True)
plt.show()
```

```python
# Fazer uma previsão
previsao = arvore.predict([[0,1,1,1]])
print("Previsão:", previsao)

# Obter probabilidades das classes
probas = arvore.predict_proba([[1,1,1,1]])[0]
print(pd.Series(probas, index=arvore.classes_))
```
